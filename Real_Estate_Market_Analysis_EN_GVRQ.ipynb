# Exploratory Data Analysis(EDA) on Residential Properties

The **(*NDA*)** platform is going to create an automated system for tracking anomalies and fraudulent activities. Therefore it is necessary to set parameters for determining the market value of real estate objects for further automated system's creation.

Using the data of the **(*NDA*)** real estate listing platform to determine the market value of real estate objects and typical parameters of apartments.

We have at our disposal a dataset of the real estate listings archive over apartments sold in St. Petersburg and other cities in the Leningrad Region for the past several years.

There are two types of data available for each real estate listing:
1. Data entered by the user.
2. Data obtained automatically based on map data (*distance to the city center, airport, nearest park and body of water*).

We create a list of settlements with their median time to the airport.

```python
local_airports = df.pivot_table(index='locality_name', values='airports_nearest', aggfunc='median')
locality_dict=local_airports.index.unique()
df['airports_nearest_new'] = df['airports_nearest']
```

```python
for locality in locality_dict:
    median_airport = df[df['locality_name']==locality]['airports_nearest'].median()
    df.loc[(df['airports_nearest_new'].isnull()) & 
                   (df['locality_name'] == locality), 'airports_nearest_new'] = median_airport
df['airports_nearest'] = df['airports_nearest_new']

#drop airports_nearest_new column
df.drop(['airports_nearest_new'], axis='columns', inplace=True)
```
